```python
# joining arrays

import numpy as np
a = np.array([[1,2],[3,4]])

print('First array:')
print(a)

b = np.array([[5,6],[7,8]])
print('Second array:')
print(b)
print('\n')
print(np.concatenate((a,b)))
print('\n')
print(np.concatenate((a,b), axis=1))
```
